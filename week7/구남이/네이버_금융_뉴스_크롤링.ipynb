# 페이지 끝까지 제목 크롤링

```python
import requests
from bs4 import BeautifulSoup
```

html 요소 확인

<dd class="articleSubject">
	<a href="/news/news_read.naver?article_id=0005267159&amp;office_id=018&amp;mode=LSS2D&amp;type=0&amp;section_id=101&amp;section_id2=258&amp;section_id3=&amp;date=20220714&amp;page=1" title="한눈에 보는 월가 투자의견...줌 목표가 올렸지만">한눈에 보는 월가 투자의견...줌 목표가 올렸지만</a>
	</dd>


```python
url = 'https://finance.naver.com/news/news_list.naver?mode=LSS2D&section_id=101&section_id2=258'

# 페이지 범위를 크게 지정해서 해당 페이지에 뉴스 기사가 존재하지 않으면 멈추도록 함

page = 0
news_titles = []
   
while True:
    page += 1
    params = { 'page' : page }
    resp = requests.get(url, params)
    html = BeautifulSoup(resp.content, 'lxml')
    # articleSubject의 a 태그에 title이 존재
    titles_html = html.select('.articleSubject > a ')
    
    if len(titles_html) != 0:
        print(page, '페이지 크롤링 시작')
        for i in range(len(titles_html)):
            news_titles.append(titles_html[i].text)
            print(i+1, titles_html[i].text)
        print(page, '페이지 크롤링 완료', '\n')

    else :
        print('\n \n',page-1, '이 마지막 페이지 입니다.')
        break 
    
```

```python
len(news_titles)
```

리스트를 텍스트 파일로 저장

```python
with open('news_titles.txt','w',encoding='UTF-8') as f:
    for title in news_titles:
        f.write(title+'\n')
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
from konlpy.tag import Okt
from PIL import Image
import numpy as np
```

저장한 텍스트 파일로 워드클라우드 생성

```python
with open('news_titles.txt', 'r', encoding='utf-8') as f:
    text = f.read()

okt = Okt()
nouns = okt.nouns(text) # 명사만 추출

words = [n for n in nouns if len(n) > 1] # 단어의 길이가 1개인 것은 제외

c = Counter(words) # 위에서 얻은 words를 처리하여 단어별 빈도수 형태의 딕셔너리 데이터를 구함

wc = WordCloud(font_path='malgun', background_color='white', colormap = 'Spectral',
               width=600, height=600, scale=2.0, max_font_size=250)
gen = wc.generate_from_frequencies(c)
plt.figure()
plt.imshow(gen)
plt.savefig("news_wordcloud.png",dpi=300)
```

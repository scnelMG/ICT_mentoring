# 거래량 상위 TOP 10

```python
import pandas as pd
import requests
URL = f"https://finance.naver.com/sise/sise_quant.naver"
r = requests.get(URL)
top_vol = pd.read_html(r.text,  encoding = 'utf-8')[1]
```

```python
top_vol.head(10)
```

```python
# 거래량 순위 데이터 프레임에서 종목명이 nan 값인 행 삭제
na_index = top_vol[top_vol['종목명'].isna()].index
top_vol.drop(na_index,axis=0, inplace=True)
# 칼럼명 N 을 순위로 바꾸기
top_vol.rename(columns = {'N' : '순위'}, inplace = True)
```

```python
top_vol
```

```python
# 거래 상위량 TOP 10
top_10_vol = top_vol[:10]
top_10_vol
```

```python
# 인덱스 지우고 데이터 프레임을 사진으로 저장하기

top_10_vol.set_index(keys='순위', drop=True, inplace=True)
dfi.export(top_10_vol, 'top_10_vol.png', max_cols=-1, max_rows=-1)
```
